<h3>Functions</h3>

```python
# Gets the name of the category
def get_category_type(row):
    try:
        categories_list = row['Category']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

<h3>FourSquare API on all neighbourhoods</h3>

```python
explore_df_list = []

for i, nbd_name in enumerate(table['Neighborhood']):
        
    try :         
        ### Getting the data of neighbourhood
        nbd_name = table.loc[i, 'Neighborhood']
        nbd_lat = table.loc[i, 'Latitude']
        nbd_lng = table.loc[i, 'Longitude']

        radius = 500 # Setting the radius as 500 metres
        LIMIT = 100 # Getting the top 100 venues

        url = 'https://api.foursquare.com/v2/venues/explore?client_id={} \
        &client_secret={}&ll={},{}&v={}&radius={}&limit={}'\
        .format(CLIENT_ID, CLIENT_SECRET, nbd_lat, nbd_lng, VERSION, radius, LIMIT)

        results = json.loads(requests.get(url).text)
        results = results['response']['groups'][0]['items']
        
        print(results)

        nearby = json_normalize(results) # Flattens JSON

        # Filtering the columns
        filtered_columns = ['venue.name', 'venue.categories', 'venue.location.lat', 'venue.location.lng']
        nearby = nearby.loc[:, filtered_columns]

        # Renaming the columns
        columns = ['Name', 'Category', 'Latitude', 'Longitude']
        nearby.columns = columns

        # Gets the categories
        nearby['Category'] = nearby.apply(get_category_type, axis=1)

        # Gets the data required
        for i, name in enumerate(nearby['Name']):
            explore_df_list.append([nbd_name, nbd_lat, nbd_lng] + nearby.loc[i, :].values.tolist())
    
    except Exception as e:
        pass
```

<h3>Complete Dataframe</h3>

```python
explore_df = pd.DataFrame([item for item in explore_df_list])
explore_df.columns = ['Neighborhood', 'Neighborhood Latitude', 'Neighborhood Longitude', 'Venue Name', 'Venue Category', 'Venue Latitude', 'Venue Longitude']
explore_df.head()
```

```python
# Preprocessing
table_onehot = pd.get_dummies(explore_df[['Venue Category']], prefix="", prefix_sep="")
table_onehot['Neighborhood'] = explore_df['Neighborhood'] 
fixed_columns = [table_onehot.columns[-1]] + list(table_onehot.columns[:-1])
table_onehot = table_onehot[fixed_columns]
table_onehot = table_onehot.groupby('Neighborhood').mean().reset_index()

table = table_onehot

# top 5 venues
num_top_venues = 15
indicators = ['st', 'nd', 'rd']

# Create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# Create a new dataframe
table_venues= pd.DataFrame(columns=columns)
table_venues['Neighborhood'] = table['Neighborhood']

for ind in np.arange(table.shape[0]):
    table_venues.iloc[ind, 1:] = return_most_common_venues(table.iloc[ind, :], num_top_venues)

table_venues.head()
```

<h3>K-Means Clustering</h3>

```python
# Set number of clusters
kclusters = 5
clustering = table.drop('Neighborhood', 1)

# Run k-means clustering
kmeans = KMeans(n_clusters = kclusters, random_state = 0).fit(clustering)

# Check cluster labels generated for each row in the dataframe
kmeans.labels_[0:5] 

# Add clustering labels
table_venues.insert(0, 'Clusters', kmeans.labels_)
```

<h3>All Data</h3>

```python
table_all = pd.read_csv("Toronto_LatLong.csv").drop('Unnamed: 0', axis=1)
table_all = table_all.join(table_venues.set_index('Neighborhood'), on='Neighborhood')
table_all.dropna(inplace = True)
table_all['Clusters'] = table_all['Clusters'].astype(int)
table_all.head()
```

<h3>Clusters</h3>

```python
map_clusters = folium.Map(location=[tor_lat, tor_lng], zoom_start=11)

x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# Add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(table_all['Latitude'], table_all['Longitude'], table_all['Neighborhood'], table_all['Clusters']):
    label = folium.Popup(str(poi) + ' (Cluster ' + str(cluster) + ')', parse_html=True)
    map_clusters.add_child(
        folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7))
       
map_clusters
```
